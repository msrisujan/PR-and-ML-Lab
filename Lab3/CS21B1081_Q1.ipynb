```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("iris.csv").drop("Id", axis=1)
data
```

```python
classes = data['Species'].unique()
train_data = pd.DataFrame()
test_data = pd.DataFrame()

for i in classes:
    classdf = data[data['Species'] == i]
    sample = classdf.sample(n=1)
    test_data = pd.concat([test_data, classdf.sample(n=1)])
    train_data = pd.concat([train_data, classdf.drop(sample.index)])


train_data
```

```python
test_data
```

```python
train_data_list = []

for i in classes:
    x = train_data[train_data['Species'] == i]
    train_data_list.append(x.iloc[:, :-1])

X_test = test_data.iloc[:, :-1]  
y_test = test_data.iloc[:, -1]  

X_test
```

```python
def Mahalanobis_distance(x,X_train):
    mean = X_train.mean(axis=0).values
    cov = X_train.cov().values
    return np.sqrt((x-mean).T @ np.linalg.inv(cov) @ (x-mean))
```

```python
pred = []
min_dist = []

for i in X_test.values:
    dist = []
    for j in train_data_list:
        dist.append(Mahalanobis_distance(i, j))
    min_dist.append(dist[np.argmin(dist)])
    pred.append(classes[np.argmin(dist)])
```

```python
compare = pd.DataFrame({'Actual': y_test, 'Predicted': pred, 'Mahalanobis Distance': min_dist})
compare
```
